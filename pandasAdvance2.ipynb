```python
import pandas as pd
import requests

# Replace with your Google Sheets URL
google_sheet_url = "https://docs.google.com/spreadsheets/d/1XJWhtAT0YH_g8aavhkfKI7w96g_9cENG/export?format=csv&gid=1242802914"

# Fetch the CSV content using requests
response = requests.get(google_sheet_url)
print(response)
# Check if the request was successful
if response.status_code == 200:
    # Read the CSV content using pandas
    df = pd.read_excel(response.content)
    print(df)
else:
    print("Failed to fetch data")

```
